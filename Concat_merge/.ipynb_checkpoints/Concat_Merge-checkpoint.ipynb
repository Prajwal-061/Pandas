```python
import pandas as pd
india_weather=pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weather
```

```python
us_weather=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
# concating the two dataframes
df=pd.concat([india_weather,us_weather],keys =['india','US'])
df
```

```python

df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
temperature_df=pd.DataFrame({
    "city":['mumbai','delhi','banglore'],
    "temperature":[32,45,30],
})
temperature_df
```

```python
windspee_df=pd.DataFrame({
    "city":['delhi','mumbai'],
    'windspeed':[7,12],
},index=[1,0])
windspee_df
```

```python
# concating two data frames by column wise
pd.concat([temperature_df,windspee_df],axis=1)
```

```python
# merge function.
df1=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
})
df1
    
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","orlando"],
    "humidity":[65,68,75],
})
df2
```

```python
# merge function
pd.merge(df1,df2,on="city")
```
