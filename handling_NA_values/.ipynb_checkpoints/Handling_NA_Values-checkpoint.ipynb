```python
import pandas as pd
# changing the datatype of day column as timestamp from string
df=pd.read_csv("weather_data.csv",parse_dates=["day"])
print(df)
print(type(df.day[0]))
```

```python
# setting the day column as index
df.set_index('day',inplace=True)
df
```

```python
# handling na values
df.fillna(0)
df
```

```python
print(df.temperature.mean())
# it puts the mean value when there is nan value
df.fillna({
    "temperature":df.temperature.mean(),
    "windspeed":df.windspeed.mean(),
    "event":"No event"
})
```

```python
# other method ffill forwards the previous value to the nan when it encounter nan.
print(df)
df.fillna(method="ffill")
```

```python
# back fill
df.fillna(method="bfill",axis='columns')
```

```python
print(df)
```

```python
# equally increments the value and puts in the na value
df.interpolate()
```

```python
# drops the rows havin na value
df.dropna()
```

```python
# only drops the rows that have all navalues
df.dropna(how="all")
```

```python
print(df)
# drops the rows if there is less than 2 non NA values
df.dropna(thresh=2)
```
