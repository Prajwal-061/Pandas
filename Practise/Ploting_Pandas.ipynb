```python
import pandas as pd
import matplotlib.pyplot as plt
air_quality=pd.read_csv("air_quality_no2.csv",index_col=0, parse_dates=True)
print(air_quality.head())
```

```python
air_quality.plot()
plt.show()
```

```python
# to plot the specific columns
air_quality["station_paris"].plot()
```

```python
#box plot
air_quality.plot.box()
```

```python
#scattering plot
air_quality.plot.scatter(x="station_paris",y="station_london",alpha=0.5)
```

```python
# i want each columns in a separate subplots
axs = air_quality.plot.area(figsize=(12, 5), subplots=True)
plt.show()
```

```python
#creating new columns from existing column
air_quality["london_mg_per_cubic"]=air_quality["station_london"]*1.8
air_quality.head()
```

```python
# other example of creating new columns from existing columns
air_quality["ratio_paris_antwerp"]=(air_quality["station_paris"] / air_quality["station_antwerp"])
air_quality.head()
                                    
```

```python
# renaming the columns name
air_quality_renamed=air_quality.rename(columns={
                                            "station_antwerp":"ABCDEF",
                                            "station_paris": "FR04014",
                                            "station_london": "London Westminster",
    })
print(air_quality_renamed.head())
type(air_quality_renamed)
```
