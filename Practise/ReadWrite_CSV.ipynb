```python
movies=pd.read_excel("movies_db.xlsx","movies")
movies.head(5)
```

```python
financial=pd.read_excel("movies_db.xlsx","financials")
financial.head(4)
```

```python
# want to conver "$$" and "Dollars" as USD using custom function.
def standardize_currency(curr):
    if curr =="$$" or curr =="Dollars":
        return "USD"
    return curr
financial=pd.read_excel("movies_db.xlsx","financials", converters={
    'currency':standardize_currency})
financial.head(4)
```

```python
df_merged=pd.merge(movies,financial,on="movie_id")
df_merged.head()
```

```python
#Merging the excel file.
df_merged.to_excel("movies_merged.xlsx",sheet_name="merged",index=False)
```

```python
weather_data = {
    "day": ["2025-08-01", "2025-08-02", "2025-08-03", "2025-08-04"],
    "temperature": [30, 32, 31, 29],
    "event": ["Sunny", "Rain", "Cloudy", "Sunny"]
}
weather=pd.DataFrame(weather_data)
weather
```

```python
stock_data = {
    "stockcode": ["AAPL", "GOOGL", "TSLA", "MSFT"],
    "price": [190.25, 2740.30, 688.50, 305.10],
    "eps": [5.11, 89.34, 3.58, 8.05]
}

stock = pd.DataFrame(stock_data)
stock
```

```python
with pd.ExcelWriter("stock_weather.xlsx") as writer:
    stock.to_excel(writer,sheet_name="stocks")
    weather.to_excel(writer,sheet_name="weather")
    
```
