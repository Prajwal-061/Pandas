```python
import pandas as pd
df=pd.read_csv("weather_by_cities.csv")
df
```

```python
df[df.city=='new york'].temperature.max()
```

```python
# groupby groups the dataframe accordint eh column. here accordint to city. there are three city
#so it create three different dataframes and the key is city name.
g=df.groupby("city")
g
```

```python
for city,data in g:
    print("city",city)
    print("data",data)
```

```python
g.get_group("mumbai") # it will give the dataframe of city mumbai
```

```python
g.get_group("paris") # it will give the dataframe of city paris
```

```python
g.max() # it will go through each dataframes and calculate the max value for each numerical columns.
```

```python
g.describe()
```

```python
g.size()
```

```python
titanic=pd.read_csv("Titanic-Dataset.csv")
titanic.head()
```

```python
# it will group on the basis of sex column and only displays the age and sex column
g=titanic[["Age","Sex"]].groupby("Sex")
g
```

```python
# it only displays the age and sex column
g.get_group("male")
```

```python
mean=titanic[["Age","Sex"]].groupby("Sex").mean()
mean
```

```python
grouped = titanic.groupby(["Sex", "Pclass"])

df1 = grouped.get_group(("female", 1))
df2 = grouped.get_group(("female", 2))
df3 = grouped.get_group(("female", 3))
df4 = grouped.get_group(("male", 1))
df5 = grouped.get_group(("male", 2))
df6 = grouped.get_group(("male", 3))
df1

```
